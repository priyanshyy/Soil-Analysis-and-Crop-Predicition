```python
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest
x.columns
```

### Spliting the data

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
acc = []
model = []
```

### Model training

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
#from sklearn.neighbors import KNeighborsClassifier
#from sklearn.naive_bayes import GaussianNB  
```

```python
dtc = DecisionTreeClassifier(criterion="entropy",random_state=2,max_depth=5)
svc = SVC()
rfc = RandomForestClassifier()
#knn = KNeighborsClassifier(n_neighbors=5)
#nb = GaussianNB() 
```

```python
model.append('Decision Tree Classifier')
model_dtc = dtc.fit(x_train,y_train)
```

```python
model.append('SVC')
model_svc = svc.fit(x_train,y_train)
```

```python
model.append('Random Forest Classifier')
model_rfc = rfc.fit(x_train,y_train)
```

```python
#model.append('K Nearest Neighbours')
#model_knn = knn.fit(x_train,y_train)
```

```python
#model.append('Naive Bayes') 
#model_nb = nb.fit(x_train, y_train) 
```

### Prediction

```python
pred1 = model_dtc.predict(x_test)
pred2 = model_svc.predict(x_test)
pred3 = model_rfc.predict(x_test)
#pred4 = model_knn.predict(x_test)
#pred5 = model_nb.predict(x_test)
```

### Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
tc1 = model_dtc.score(x_train,y_train)
tc2 = model_svc.score(x_train,y_train)
tc3 = model_rfc.score(x_train,y_train)
#tc4 = model_knn.score(x_train,y_train)
#tc5 = model_nb.score(x_train,y_train)
print(tc1,
      tc2, 
      tc3)
```

```python
acc1 = accuracy_score(y_test,pred1)
acc.append(acc1)
acc1
```

```python
acc2 = accuracy_score(y_test,pred2)
acc.append(acc2)
acc2
```

```python
acc3 = accuracy_score(y_test,pred3)
acc.append(acc3)
acc3
```
